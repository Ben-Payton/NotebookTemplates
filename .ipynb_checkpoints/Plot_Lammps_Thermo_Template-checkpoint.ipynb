# Opening a file and returning it as a list:

```python
def File_to_list(file_name):
    with open(file_name) as file:
        file_list = file.readlines()
    return file_list    

def Get_Thermo_Lists(file_list):
    
    Temp_List = []
    All_Data = []
    Search_start_string = "Per MPI rank memory"
    Search_stop_string = "Loop time"
    
    for i in range(len(file_list)):
        if Search_start_string in file_list[i]:
            t_List=[]
            i=i+1
            t_List.append(file_list[i].split())
            i=i+1
            while Search_stop_string not in file_list[i]:
                #print("Hello")
                dat_list = file_list[i].split()
                #for p in range(len(t_List)):
                #    t_List[p].append(dat_list[p])
                t_List.append(dat_list)
                i=i+1
            All_Data.append(t_List)
    return All_Data

def List_to_Dict(List):
    All_data = []
    for dat_set in List:
        #print(dat_set)
        temp_dict = {}
        key_list = []
        for i in dat_set[0]:
            temp_dict[i] = []
            key_list.append(i)
        for row in dat_set[1:]:
                for item in range(len(row)):
                    temp_dict[key_list[item]].append(float(row[item]))
        All_data.append(temp_dict)
            
    return All_data
    
```

```python
list_thing = File_to_list("stdout.txt")
Data_Stuff = Get_Thermo_Lists(list_thing)
Final_data = List_to_Dict(Data_Stuff)
print(Final_data[4])
```

```python
fig, axs = plt.subplots(1,3,figsize=(15,5))

x=4

axs[0].plot(Final_data[x]["Step"],Final_data[x]["Press"],color="blue")
axs[1].plot(Final_data[x]["Step"],Final_data[x]["Temp"],color="red")
axs[2].plot(Final_data[x]["Step"],Final_data[x]["Density"],color="green")
```
